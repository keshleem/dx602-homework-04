Rules:

* Then run the provided test cases and print results exactly as shown.

**Sample Output:**


````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW

def middlest_element(values):
  '''function to return the middle value of the list for both odd and even lengths'''

# variable for length of list
  list_length = len(values)

  # if list is empty, return None
  if values == []:
    return None
    
  # if the length of the string is divisible by 2 without a remainder (odd), return the middle element
  if list_length % 2 != 0:
    return values[list_length // 2]

  # else (even), return the element to the left by taking the length, subtracting 1 and then round down  
  else:
    return values[(list_length - 1) // 2]

````

````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def mean(values):
  '''return the arithmetic mean of list of numbers'''

  # define accumulator variable
  sum_values = 0

  # use a loop to iterate over the list
  for i in range(len(values)):
    # the new sum total now equals 0 + the rest of the elements in the list
    sum_values = sum_values + values[i]

  # mean is total / length of list. go through each element in list
  mean = sum_values / len(values)
  
  # return the calculated mean variable
  return mean
````

````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW

def mean(values):
  '''returns the mean'''
  
  # define an accumulator variable
  sum_mean = 0 

  # use a loop to iterate over the list
  for i in range(len(values)): 
    # the new sum total now equals 0 + the rest of the elements in the list
    sum_mean = sum_mean + values[i] 
  
  # mean is total / length of list. go through each element in list
  mean = sum_mean / len(values) 
  
  # return the calcuated mean variable
  return mean 

def variance(values):
  '''returns the population variance of a list of number using the variance formula'''

  #accumulator to start from 0 for sum of squared diff
  squared_diff = 0

  # use a loop to iterate over the list
  for i in range(len(values)):
    # variable for (x - mean) ** 2
    squared_diff = squared_diff + (values[i] - mean(values)) ** 2

  # calculates the variance of the list
  variance = squared_diff / len(values)

  # gives us the calculated variance
  return variance


  
````

**Problem 8**

**Concept:** *Accumulator pattern with numeric result, list iteration, index-based loop.*

**Task:**

Write a function `covariance(xs, ys)` that computes the population covariance between two series:

$$
\text{covariance} = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})
$$

Rules:

* If `xs` and `ys` are not the same length, print:
    
    `xs and ys must be the same length`
    
    and return `0`.

* Must use an index-based loop.

* Must use your `mean()` function.

* Then run the given test case and print exactly in the same format.

**Sample Output:**

```
x values: [6, 3, 7, 9, 5]
y values: [4, 1, 7, 5, 0]
covariance: 3.8
```


````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def mean(values):
  '''returns the mean'''
  
  # define an accumulator variable
  sum_mean = 0 

  # use a loop to iterate over the list
  for i in range(len(values)): 
    # the new sum total now equals 0 + the rest of the elements in the list
    sum_mean = sum_mean + values[i] 
  
  # mean is total / length of list. go through each element in list
  mean = sum_mean / len(values) 
  
  # return the calcuated mean variable
  return mean

def covariance(xs, ys):
  '''calculates the covariance'''

  #accumulators for element differences and means
  mean_xs = 0
  mean_ys = 0
  cal_xy = 0

  # if the length of xs and ys are not the same, return 0 for covariance
  if len(xs) != len(ys):
    return 0

  # get the mean of xs and ys by calling mean(values)
  mean_xs = mean(xs)
  mean_ys = mean(ys)

  # use a loop to iterate over the list. Since we want lengths the same, can use xs for both xs and ys
  for i in range(len(xs)):
  # go through each list, go to each element, subtract from mean of that element, then multiply with other list
    cal_xy = cal_xy + (xs[i] - mean_xs) * (ys[i] - mean_ys)

  # dividing the multiplcation of the xs and ys elements by length to get the covariance
  covariance = cal_xy / len(xs)

  # return the calculated covariance
  return covariance
````
